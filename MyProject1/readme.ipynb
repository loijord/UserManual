# Kaip naudotis Jupyter Notebook aplinka
Matematikos pamokose dalį laiko ketinu per projektorių rodyti užrašus, iliustracijas, animacijas, interakcijas. Visa medžiaga talpinama internete. Jei norite išbandyti programavimą Python kalba arba pramokti efektyviai rinkti matematinį tekstą kompiuteriu, kviečiu susipažinti su tam skirtu Jupyter Notebook įrankiu. Norintiems bendradarbiauti ši dokumentacija padės ne tik naudotis įkelta medžiaga, bet ir ją papildyti bei teikti pasiūlymus. Turinys padalintas į dvi dalis: 

* Instrukcijas nekantriems, kurie nori išbandyti Jupyter Notebook
* Instrukcijas, norintiems susipažinti su Jupyter Notebook nuodugniai: Conda aplinkų valdymu, Python paketų instaliavimu, atnaujinimu ir trynimu, projektų valdymu GitHub puslapyje.

## Instrukcijos nekantriems 
### Atsisiuntimas ir Python paleidimas

* Atsisiųskite [Anaconda](https://www.anaconda.com/products/individual)
* Suinstaliuokite Anaconda (uždėsite varnelę ant Register Anaconda3 as my default Python 3.8) - reikės 3GB vietos, gali tekti palaukti 2 - 45 min.
* Išbandykite, ar veikia Python: 
  * Iš Start Menu paleiskite *Anaconda Prompt*, iššokusiame lange suveskite ,,python''. Turėtų atsidaryti Python interpretatorius.
  * Įsitikite, kad komandos veikia. Suveskite `print('Hello, world!')`. Turėtų lange išspausdinti ***Hello, world!*** Tai buvo jūsų pirmoji komanda.
  * Paspauskite CTRL+Z, kad grįžtumėte iš Python interpretatoriaus į Anaconda Prompt.

### Conda virtuali aplinka

Iš pradžių užteks tik šios komandos:

`conda install jupyter matplotlib numpy`

Suveskite šią komandą *Anaconda Prompt* terminale. Ji padės suinstaliuoti pagrindinius Python ir Conda paketus, kuriuos naudosite (kol kas tik) vienoje pagrindinėje aplinkoje. `jupyter` paketas ypatingas: jis užtikrina, kad galėtumėte naudotis Jupyter Notebook. Visi kiti paketai - tai tam tikros Python kalbos bibliotekos.

### Programavimas

Jeigu reikia atlikti paprastą kelių eilučių skaičiavimą, tam Jupyter Notebook nebūtinas. Galime įvesti komandą `python` ir programuoti per terminalą. Arba net gi tą daryti [internete](http://pythontutor.com)

Pavyzdžiui mums reikia apskaičiuoti, kokio ilgio kelią nukrenta kūnas per pirmąsias 5 sekundes. Duota, kad per 1 sekundę jis nukrenta 4.9 m. Remiamės formule $s=\frac{gt^2}{2}$. Iš pradžių patikriname ar formulė veikia atveju $t=1$. Po to taikome ją atveju $t=5$.

    >>> g=9.8
    >>> t=1
    >>> g*t**2/2
    4.9
    >>> t=5
    >>> g*t**2/2
    122.50000000000001

Atsakymas - 122.5 metro. Matome, kad Python gali padėti atlikti paprastus skaičiavimus.

### Jupyter Notebook

Kuo naudinga ankstesnė programa? Greičiausiai padėjo įsitikinti, kad formulė veikia, bet kažin, ar padės išsaugoti darbo rezultatus. Būtų gerai aprašyti, kas nuveikta ir išsisaugoti. Tam ypač patogus *Jupyter Notebook*. Uždarykime *Python* su CTRL+C ir įveskim į *Anaconda Prompt* komandą `jupyter notebook`. Atsidariusioje aplikacijoje turi matytis failai, kuriuos mato Conda aplinka. Čia galite kurti naujus *Jupyter Notebook* (`.ipynb` formato) failus ir aplankus.

Peržiūrėkite, kaip galėtų atrodyti tvarkingai aprašytas failas [`kritimas.ipynb`](kritimas.ipynb). Jei norite, galite pamėginti surinkti tokį patį tekstą su tokiomis pačiomis komandomis. Į kiekvieną langelį tekstas vedamas dviem režimais:

* `Markdown`, jei reikia sukurti užrašus
* `Code`, jei reikia rašyti ir vykdyti kodą

Kiekvieną langelį galima paleisti su SHIFT + ENTER. Suvestas tekstas:

    [Markdown]: 
    Apskaičiuosime, kokio ilgio kelią nukrenta kūnas per pirmąsias 5 sekundes. Remsimės formule $s=\frac{gt^2}{2}$. Laikysime, kad $g=9.8$
    [Markdown]: 
    Kai $t=1$, tai gauname, kad $s=\frac{9.8 \cdot 1^2}{2}=4.9$:
    [Code]: 
    g=9.8
    t=1
    g*t**2/2 #taip skaičiuojama gt²/2 reikšmė
    [Markdown]: 
    Kai $t=5$, tai gauname, kad $s=\frac{9.8 \cdot 5^2}{2}=122.5$:
    [Code]:
    t=5
    g*t**2/2
    [Markdown]:
    Tą patį galima padaryti susikuriant funkciją:
    [Code]:
    def s(t, g=9.8):
        return g*t**2/2
    s(1), s(5)
    
Daugiau pavyzdžių, kaip įvesti matematines formules, galima rasti [matematikos forume](https://www.ematematikas.lt/forumas/matematiniai-simboliai-latex-sintakse-t1167.html)

Štai sudėtingesnio failo [`kvadratinės_funkcijos.ipynb` pavyzdys](kvadratinės_funkcijos.ipynb). 

Tikiuosi, kad šios instrukcijos padėjo susipažinti su *Jupyter Notebook* įrankio naudojimu užrašams ir programavimui. Norėdami uždaryti šią programą galite išjungti visus langus ir terminale įvesti CTRL + C.

## Instrukcijos norintiems susipažinti nuodugniai

### Plačiau apie Conda aplinkas

Atidarius *Jupyter Notebook* rekomenduojame susikurti aplanką, skirtą jūsų programoms laikyti. Geriausia būtų turėti aplanką, skirtą visiems projektams ir jame susikurti kitą aplanką, skirtą pirmajam projektui:

  * New -> Folder
  * Uždėkite varnelę ant Untitled Folder, paspauskite Rename ir pervardinkite aplanką į my_jupyters
  * Įeikite į šį aplanką (jis turi būti tuščias)
  * Panašiai šiame aplanke susikurkite kitą aplanką MyProject1 ir įeikite į jį.
  * Aplanke MyProject1 sukurkite naują *Jupyter Notebook* failą (New -> Python). 
  * Pabandykite patys sukurti anksčiau minėtą failą `kritimas.ipynb`, kad galėtume ištestuoti projektų kėlimą į *GitHub*

Failui `kritimas.ipynb` nereikia jokių papildomų bibliotekų. Tačiau sudėtingesnėse programose jos neišvengiamos. Pavyzdžiui norint sukurti funkcijų braižyklę gali prireikti `numpy` (masyvų operacijų) ir `matplotlib` (braižymo) paketų. Norint automatiškai prastinti reiškinius reikia jau sudėtingesnio `sympy` (simbolinių skaičiavimo) paketo. O štai populiariausias matematikos mokymo kanalas [`3blue1brown`](https://www.youtube.com/channel/UCYO_jab_esuFRV4b17AJtAw) savo animacijoms kurti specialiai parašė `manim` paketą. Visi šie paketai užima daug vietos, todėl projektams verta naudoti skirtingas virtualias aplinkas - specifines direktorijas, kuriose laikomi tik konkrečiam projektui būtini bibliotekų rinkiniai.

Dėl patogumo visus dažniausiai naudojamus paketus atsisiųsime į pagrindinę Conda aplinką, pavadintą `base`, o sudėtingesniems projektamskuriamos kitos Conda aplinkos. Vienas iš būdų tvarkyti šias aplinkas yra naudoti *Anaconda Prompt* terminalą. Išvardysiu sąrašą naudingų komandų:

|Komanda|Kas vykdoma|
|:---|:---|
|`cd MyFolder`|Vartotojo kelyje nueina į MyFolder aplanką (jei tokį randa)|
|`cd..`|Vartotojo kelyje žengia žingsnį atgal|
|`conda info --envs`|Parodo informaciją apie visas naudojamas aplinkas (žvaigždutė nurodo, kuri aplinka aktyvi)|
|`conda update conda`|Atnaujina Conda versiją į naujausią|
|`conda create --name my_playground`|Sukuria naują aplinką pavadinimu my_playground|
|`conda activate my_playground`|Aktyvuoja aplinką pavadinimu my_playground|
|`conda install jupyter matplotlib*`|Aktyvioje aplinkoje suinstaliuoja jupyter ir matplotlib paketus|
|`conda update matplotlib`|Atnaujinti tam tikrą aktyvios aplinkos paketą|
|`conda update --all`|Atnaujina visas aktyvios aplinkos bibliotekas|
|`conda deactivate`|Deaktyvuoja aplinką (`base` aplinka tampa aktyvia)|
|`conda remove --name my_playground -all`|Ištrina aplinką ir visus jos paketus (prieš tai reikia deaktyvuoti)|

*Jupyter instaliacija yra būtina tam, kad galėtume patogiai rašyti kodą

Iš pradžių pakanka `base` aplinkoje susiinstaliuoti tik pagrindinius paketus:

`conda install jupyter matplotlib numpy`

Vėliau pagal poreikį bibliotekų sąrašą galima praplėsti. Jei reikia daug paketų, nepatartina naudoti tos pačios aplinkos visiems projektams, nes paketai gali būti nesuderinti tarpusavyje (remtis tuo pačiu, bet skirtingų versijų paketu). Faile [`kvadratinės_funkcijos.ipynb`](kvadratinės_funkcijos.ipynb) naudojami `matplotlib` ir `numpy` paketai pilnai gali būti pajungti iš  `base` aplinkos.

## Github

Natūralu, kad programų kodai išsimėto, užsimiršta. Vienas iš būdų su tuo susitvarkyti - rašyti tvarkingą kodą, jį komentuoti. 
Tačiau tai neapdraudžia nuo kompiuterio gedimų, kai failai prarandami. Tuo labiau, jokie programuotojai šiais laikais kodo nepersiuntinėja per *Gmail* ar *Facebook*. Tam yra skirta versijų kontrolės platforma *GitHub*, leidžianti projektus talpinti internete, juos nuolat atnaujinti ar parsisiųsti kitus. Kaip įprasta programuotojams, tai turi būti daroma tam tikrų komandų pagalba sugaištant kuo mažiau laiko. Tam, kad galėtume parsisiųsti ir įsikelti *Jupyter Notebook* failus, reikės:

  * Atsisiųsti ir suinstaliuoti [`git`](https://git-scm.com/downloads) (Reikės 260Mb vietos)
  * Susikurti paskyrą [`GitHub`](https://github.com)
  
Toliau vietą, kur saugomas kodas *GitHub* vadinsime repozitorija (*repository*). 
[Toliau](https://towardsdatascience.com/how-to-use-git-github-with-jupyter-notebook-7144d6577b44)

### Projekto parsisiuntimas

Pati paprasčiausia versijų kontrolės komanda yra `git clone`. Ji leidžia parsisiųsti projektus iš *GitHub* į jūsų kompiuterį. Projektas bus laikomas tam tikrame aplanke. *Anaconda Prompt* terminale naudojantis komandomis `cd` ir `cd..` išsirenkame vietą, kur norėsime saugoti atsisiųstą projektą. Geriausia būtų ten pat, kur ir jau esamas `MyProject1`, t.y. reikėtų nurodyti kelią iki `my_jupyters` aplanko. Dabar galite pabandyti atsisiųsti naują projektą `projectA` įvedę

`git clone https://github.com/amit1rrr/projectA.git`

Jei jūsų kompiuteryje šalia `MyProject1` atsirado projekto `projectA` kopija, komanda veikia puikiai.

### Projekto įkėlimas

Naujo projekto įkėlimas į GitHub - daug sudėtingesnis procesas. Norint įkelti naują projektą reikės GitHub paskyros vartotojo vardo, slaptažodžio, projekto pavadinimo jūsų kompiuteryje ir pavadinimo, kuriuo bus saugomas repozitorija *GitHub*. Tarkime:

* jūsų vartotojo vardas: *nniicckknnaammee*
* įkeliamo projekto pavadinimas: *MyProject1*
* projekto pavadinimas GitHub'e: *UserManual*
* GitHub identifikacijos kodas: *giRAa8TvuseijVCXplrjUJUPS6TVVVk7xo3oqSGl*.

Talpinant projektą *GitHub*e kompiuteryje pirmiausiai sukuriama laikina *Git* repozitorija (.git/ aplankas), kurioje fiksuojami (`git commit`) visi pakeitimai. Tik užfiksavus pakeitimus projektą galima perkelti (`git push`) iš laikinos *Git* repositorijos į *GitHub* repozitoriją. Pirmą kartą jungiantis prie *Git* reikėtų prisistatyti naudojant savo GitHub vartotojo vardą ir paštą, kad *Git*'as galėtų atskirti, į kurią *GitHub* paskyrą norite kelti failus. Išsaugokite pakeitimus, išeikite iš *Jupyter Notebook* ir į *Anaconda Prompt* įveskite tokias komandas:

    git config --global user.name "nniicckknnaammee"
    git config --global user.email "email@example.com"

Tada norint įkelti naują projektą reikės įvesti tokias komandas:

    git init
    git add MyProject1
    git commit -m "Add my first commit message"
    curl -H "Authorization: token giRAa8TvuseijVCXplrjUJUPS6TVVVk7xo3oqSGl" https://api.github.com/user/repos -d "{\"name\": \"User
    Manual\"}"
    git remote set-url origin git@github.com:nniicckknnaammee/UserManual.git
    git push origin master

Jei norite įgalinti kopijavimą į *Anaconda Prompt* terminalą, spauskite dešinį pelės klavišą ir uždėkitę varnelę prie *Options -> Edit Options -> QuickEdit Mode*

|Komanda|Kas vykdoma|
|:---|:---|
|`git init`|Sukuriama laikina *Git* repozitorija|
|`git add MyProject1`|Pridedamas projektas MyProject1, kurio pakeitimai bus fiksuojami|
|`git commit -m "Add my first commit message"`|Visų pridėtų failų pakeitimų fiksavimas; būtina įrašyti, kas pakeista|
|`curl -H "Authorization: token giRAa8TvuseijVCXplrjUJUPS6TVVVk7xo3oqSGl" https://api.github.com/user/repos -d "{\"name\": \"UserManual\"}"`|Automatiniu būdu *GitHub* sukuriama repozitorija naudojant identifikaciją (token); Identifikaciją reikia [susikurti GitHube](https://docs.github.com/en/github/authenticating-to-github/keeping-your-account-and-data-secure/creating-a-personal-access-token)|
|`git remote set-url origin git@github.com:nniicckknnaammee/UserManual.git`|Parenkama GitHubo repozitorija, kur bus keliamas projektas|
|`git push origin master`|Įkeliamas projektas|

Komanda su `curl` nebūtina. Vietoj jos naują GitHub repozitoriją galite sukurti ir rankiniu būdu [prisijungę](https://github.com/new) prie savo *GitHub* paskyros.

Greičiausiai pastebėjote, kad naudojantis *Anaconda Prompt* keliant failus būtina nutraukti darbą *Jupyter Notebook* aplinkoje. Yra paprastesnė išeitis šalia pasijungti ir kitą terminalą *Command Prompt*, esantį visuose kompiuteriuose. Tuomet pakaks tik išsaugoti pakeitimus ir nenutraukiant darbo tas pačias `git` komandas įvesti į *Command Prompt*. 

### Projekto atnaujinimas
Kai jau repozitorija sukurta, užtenka užfiksuoti ir įkelti projekto atnaujinimus:

    git add MyProject1/kritimas.ipynb
    git commit -m "Add my first commit message"
    git push origin master
    
[Šioje nuorodoje](https://www.geeksforgeeks.org/how-to-upload-project-on-github-from-jupyter-notebook/) rasite šiek tiek išsamesnį paaiškinimą, kaip naudotis *Jupyter Notebook* langu ir talpinti projektus GitHube.

Instrukcijos, kaip prižiūrėti projektą, kai su juo dirba keli autoriai, lieka už šio teksto ribų. Tam reikia `git pull`, `git merge` ir kitų komandų. Apie jas galite pasiskaityti kitame [išsamiame aprašyme](https://towardsdatascience.com/how-to-use-git-github-with-jupyter-notebook-7144d6577b44).
